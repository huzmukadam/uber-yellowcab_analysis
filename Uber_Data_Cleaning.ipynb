### Convert the date and time information in the data frame from a string to a datetime object. It is more convenient to work with datetime objects.

```python
y = []
for x in df500['Date/Time']:
    y.append(dt.strptime(x, '%m/%d/%Y %H:%M:%S'))

print(type(y[0]))
```

```python
df500['datetime'] = y
```

```python
df500.head()
```

```python
type(df500['datetime'][5])
```

### Convert to datetime for full dataframe 'df'

```python
datetime_list = []
for x in df['Date/Time']:
    datetime_list.append(dt.strptime(x, '%m/%d/%Y %H:%M:%S'))

print(type(datetime_list[0]))
```

```python
df['datetime'] = datetime_list
```

```python
df.drop(['Date/Time'], axis=1, inplace = True)
```

```python
df.head()
```

```python
df.to_pickle('uber_clean_data.pkl') #Save cleaned data 
```

### Clean up lyft data

```python
df_lyft = pd.read_csv('other-Lyft_B02510.csv')
```

```python
df_lyft.head()
```

```python
datetime_list = []
for x in df_lyft['time_of_trip']:
    datetime_list.append(dt.strptime(x, '%m/%d/%Y %H:%M'))

print(type(datetime_list[0]))
```

```python
df_lyft['datetime'] = datetime_list
```

```python
df_lyft.drop(['time_of_trip'], axis=1, inplace = True)
```

```python
df_lyft.head()
```

```python
df_lyft.sort_values('datetime', inplace = True)
```

```python
df_lyft.drop('Unnamed: 3', axis = 1, inplace = True)
```

### Convert pickup and dropoff datetime to datetime object. 

```python
pickup_datetime_list = []
for x in df_yellow['pickup_datetime']:
    pickup_datetime_list.append(dt.strptime(x, '%Y-%m-%d %H:%M:%S'))

print(type(pickup_datetime_list[0]))
```

```python
df_yellow['pickup_datetime'] = pickup_datetime_list
```
